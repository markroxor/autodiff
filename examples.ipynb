```python
from autodiff import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = Variable("x")
y = Variable("y")
```

```python
zero = Constant(0)

one = Constant(1)
two = Constant(2)
three = Constant(3)

pi = Constant(np.pi)
```

```python
tan = Function.Sin(zero) / Function.Cos(zero)
tan.evaluate(), tan.grad()
```

```python
tan = Function.Tan(zero)
tan.evaluate(), tan.grad()
```

```python
exp = Function.Exp(one)
exp.evaluate(), exp.grad()
```

```python
def plot(fun):
    # plot the function
    plt.plot(t, fun.evaluate())
    
    # plt the gradient
    plt.plot(t, fun.grad())
    
    plt.grid(True)
    plt.show()
```

```python
# range of plot
t = np.arange(0.0, 1.0, 0.01)
# Sin Function
plot(Function.Sin(Constant(2*np.pi*t)))
```

```python
# range of plot
t = np.arange(0.0, 1.0, 0.01)
# Sin Function
plot(Function.Cos(Constant(2*np.pi*t)))
```

```python
# range of plot
t = np.arange(0.0, 1.0, 0.01)
# Exp function
plot(Function.Exp(Constant(2*2*np.pi*t)))
```

```python
# range of plot
t = np.arange(-0.45, 0.45, 0.01)
# Tan
plot(Function.Tan(Constant(np.pi*t)))
```

```python
# range of plot
t = np.arange(0.0, 1.0, 0.01)
# Exp function

exp = Variable('x') ** Constant(2)
exp.update({'x':t, 'y':t})
plot(exp)
```
